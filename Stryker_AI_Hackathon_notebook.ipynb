```python
sns.scatterplot(x=X["pelvic_incidence"], y=y["lumbar_lordosis_angle"])
plt.xlabel("Pelvic Incidence")
plt.ylabel("Lumbar Lordosis Angle")
plt.title("Scatter Plot with x and y from Different DataFrames")
plt.show()
```

```python
for i in X.columns:
    sns.stripplot(x=X[i],hue=y['class'])
    plt.title("Strip Plot of " + i )
    plt.xlabel("Values")
    plt.show()
```

```python
correlation = X["pelvic_incidence"].corr(X["pelvic_tilt"], method='pearson')
print(f"Pearson Correlation: {correlation:.2f}")

# Scatter plot with regression line
sns.lmplot(x="pelvic_incidence", y="pelvic_tilt", data=X, line_kws={"color": "red"})
plt.title("Scatter Plot of Pelvic Incidence vs Pelvic Tilt")
plt.xlabel("Pelvic Incidence")
plt.ylabel("Pelvic Tilt")
plt.show()
```

```python
df = pd.concat([X, y], axis=1)

# Set up the grid
fig, axes = plt.subplots(nrows=6, ncols=4, figsize=(24, 36), constrained_layout=True)
sns.set_style("whitegrid")

# Define measures of spread and central tendency for annotations
def annotate_measures(ax, column, data):
    mean = data[column].mean()
    median = data[column].median()
    std_dev = data[column].std()
    q1 = data[column].quantile(0.25)
    q3 = data[column].quantile(0.75)
    iqr = q3 - q1
    text = (
        f"Mean: {mean:.2f}\n"
        f"Median: {median:.2f}\n"
        f"Std Dev: {std_dev:.2f}\n"
        f"Q1: {q1:.2f}\n"
        f"Q3: {q3:.2f}\n"
        f"IQR: {iqr:.2f}"
    )
    ax.text(0.95, 0.95, text, transform=ax.transAxes, fontsize=9,
            verticalalignment='top', horizontalalignment='right',
            bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))

# Plotting
for i, column in enumerate(X.columns):
    # Histogram
    sns.histplot(data=df, x=column, kde=True, ax=axes[i, 0], binwidth=5)
    axes[i, 0].set_title(f"Histogram of {column}")

    # Violin Plot
    sns.violinplot(data=df, x='class', y=column, ax=axes[i, 1])
    axes[i, 1].set_title(f"Violin Plot of {column}")

    # Box Plot
    sns.boxplot(data=df, x='class', y=column, ax=axes[i, 2])
    axes[i, 2].set_title(f"Box Plot of {column}")

    # 1D Scatter Plot
    scatter_ax = axes[i, 3]
    sns.stripplot(data=df, x=column, y='class', hue='class', dodge=True, ax=scatter_ax, jitter=True)
    annotate_measures(scatter_ax, column, df)
    scatter_ax.set_title(f"1D Scatter Plot of {column}")
    scatter_ax.legend([], frameon=False)  # Remove redundant legend

# Add a global title
fig.suptitle("Visualizations of Each Feature in X Grouped by Class", fontsize=20, y=1.02)
plt.show()
```
